```python
# Let's only get the data after 1990 - 01 - 01 and before 2024 - 03 - 01

fed_interest_rates = fed_interest_rates[fed_interest_rates['Date']>= start]

volatility_index = volatility_index[volatility_index['Date']>= start ]

m2_money_supply = m2_money_supply[m2_money_supply['Date']>= start ]


fed_interest_rates = fed_interest_rates[fed_interest_rates['Date'] <= end]

volatility_index = volatility_index[volatility_index['Date'] <= end ]

m2_money_supply = m2_money_supply[m2_money_supply['Date'] <= end ]
```

Let's only get the data for the 01st of each month. 

```python
# The date set we want - we can convieneintly grab them from previous monthly dataset.

date_set = bond_yield['Date']
date_set
```

```python
interest_rates = []


for date in date_set:
    interest_rates.append(fed_interest_rates[fed_interest_rates['Date'] == date].iloc[0,1])
    
        
```

```python
fed_interest_rates_new = pd.DataFrame({'Date': date_set, 'fed_interest_rates': interest_rates})
```

```python
fed_interest_rates_new.shape
```

The shape of the interest data matches with the other data.

However, this approach cannot be used with volatility index because we dropped the holidays from volatility list and the first day of a month could be a holiday. For instance, January 01st. Therefore, for missing first day of months, we get the data from the nearest date to the first day of month.

```python
def nearest(items, pivot):
    return min(items, key=lambda x: abs(x - pivot))
```

```python
vi = []
i = 0

for date in date_set:
    
    try:
        vi.append(volatility_index[volatility_index['Date'] == date].iloc[0,1])
    except IndexError:
        vi.append(volatility_index[volatility_index['Date'] == nearest(volatility_index['Date'],date)].iloc[0,1])
```

```python
volatility_index_new = pd.DataFrame({'Date': date_set, 'volatility_index': vi})
```

```python
volatility_index_new.shape
```

The shape of the interest data matches with the other data.

For weekly data of M2 money supply we can use the same strategy of finding the value of the nearest date to the first of month. 

```python
m2 = []

for date in date_set:
    
    m2.append(m2_money_supply[m2_money_supply['Date'] == nearest(m2_money_supply['Date'],date)].iloc[0,1])
```

```python
m2_money_supply_new = pd.DataFrame({'Date': date_set, 'm2_money_supply': m2})
```

```python
m2_money_supply_new.shape
```

GDP data also exist only every quarter so let's convert them to monthly data by assuming GDP is same in the beginning of every month in a particular quarter.

```python
G = []

for date in date_set:
    
    G.append(gdp[gdp['Date'] == nearest(gdp['Date'],date)].iloc[0,1])
```

```python
gdp_new = pd.DataFrame({'Date': date_set, 'gdp': G})
```

```python
gdp_new.shape
```
